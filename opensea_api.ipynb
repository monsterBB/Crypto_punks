```python
import requests
import pandas as pd
import json, datetime
from datetime import date, timedelta, datetime
import numpy as np
from pandas.core.frame import DataFrame
```

```python
url = "https://api.opensea.io/api/v1/events"
offset = "0"
querystring =  {"asset_contract_address":"0xF87E31492Faf9A91B02Ee0dEAAd50d51d56D5d4d","event_type":"successful"
                ,"only_opensea":"false","offset":offset,"limit":"300","occurred_before":"2021-05-07T00:00:00"}
response = requests.request("GET", url, params=querystring).json()
#"offset":"0","limit":"1000000","occurred_after":"2017-01-01T00:00:00"
```

```python
print(json.dumps(response, indent =4))
```

```python
data = response['asset_events']
```

```python
token_id = []
name = []
asset_contract = []
#buyer_name = []
buyer_adress = []
#seller_name = []
seller_adress = []
payment_token = []
total_price = []
time_stamp = []
usd_exchange_rate = []
for event in data:
    try:
        token_id.append(event['asset']['id'])
    except:
        token_id.append(None)
    try:
        name.append(event['asset']['name'])
    except:
        name.append(None)
    try:
        asset_contract.append(event['asset']['asset_contract']['name'])
    except:
        asset_contract.append(None)
    try:
        buyer_adress.append(event['asset']['owner']['address'])
    except:
        buyer_adress.append(None)
    try:
        payment_token.append(event['payment_token']['symbol'])
    except:
        payment_token.append(None)
    try:
        total_price.append(event['total_price'])
    except:
        total_price.append(None)
    try:
        usd_exchange_rate.append(event['payment_token']['usd_price'])
    except:
        usd_exchange_rate.append(None)
    time_stamp.append(event['transaction']['timestamp'])
    try:
        seller_adress.append(event['seller']['address'])
    except:
        seller_adress.append(None)
```

```python
df = pd.DataFrame()
```

```python
df['name'] = name
df['token_id'] = token_id
df['asset_contract'] = asset_contract
df['payment_token'] = payment_token
df['total_price'] = total_price
df['usd_exchange_rate'] = usd_exchange_rate
df['seller_adress'] = seller_adress
df['buyer_adress'] = buyer_adress
df['time_stamp'] = time_stamp
```

```python
url = "https://api.opensea.io/api/v1/events"
Decentraland = pd.DataFrame()
for i in range(0,50):
    offset = str(300*i)
    querystring =  {"asset_contract_address":"0xF87E31492Faf9A91B02Ee0dEAAd50d51d56D5d4d","event_type":"successful"
                ,"only_opensea":"false","offset":offset,"limit":"300","occurred_before":"2021-05-07T00:00:00"}
    response = requests.request("GET", url, params=querystring).json()
    data = response['asset_events']
    token_id = []
    name = []
    asset_contract = []
    buyer_adress = []
    seller_adress = []
    payment_token = []
    total_price = []
    time_stamp = []
    usd_exchange_rate = []
    for event in data:
        try:
            token_id.append(event['asset']['id'])
        except:
            token_id.append(None)
        try:
            name.append(event['asset']['name'])
        except:
            name.append(None)
        try:
            asset_contract.append(event['asset']['asset_contract']['name'])
        except:
            asset_contract.append(None)
        try:
            buyer_adress.append(event['asset']['owner']['address'])
        except:
            buyer_adress.append(None)
        try:
            payment_token.append(event['payment_token']['symbol'])
        except:
            payment_token.append(None)
        try:
            total_price.append(event['total_price'])
        except:
            total_price.append(None)
        try:
            usd_exchange_rate.append(event['payment_token']['usd_price'])
        except:
            usd_exchange_rate.append(None)
        time_stamp.append(event['transaction']['timestamp'])
        try:
            seller_adress.append(event['seller']['address'])
        except:
            seller_adress.append(None)
    df = pd.DataFrame()
    df['name'] = name
    df['token_id'] = token_id
    df['asset_contract'] = asset_contract
    df['payment_token'] = payment_token
    df['total_price'] = total_price
    df['usd_exchange_rate'] = usd_exchange_rate
    df['seller_adress'] = seller_adress
    df['buyer_adress'] = buyer_adress
    df['time_stamp'] = time_stamp
    Decentraland = Decentraland.append(df, ignore_index=True, sort=False)
```

```python
Decentraland.to_csv('Decentraland.csv')
```
